```python
import numpy as np
import pandas as pd
```

```python
class LinearRegression():
  def __init__(self):
    self.n=15
    self.x=[130,120,166,170,155,135,170,188,150,160,164,155,168,167,169]
    self.y=[50,60,44,77,58,70,80,55,90,55,57,67,55,64,55]

  def fit(self):
      self.m=0
      self.b=0
      x_bar=np.mean(self.x)
      y_bar= np.mean(self.y)

      sum1=0
      sum2=0
      for i in range(self.n):
        sum1=sum1+(self.x[i]-x_bar)*(self.y[i]-y_bar)
        sum2=sum2+(self.x[i]-x_bar) ** 2
      self.m= sum1 / sum2
      self.b=y_bar-self.m * x_bar

  def predict(self, x):
        return self.m * x + self.b
```

```python
linear_reg = LinearRegression()
linear_reg.fit()
test_x = 165
predicted_y = linear_reg.predict(test_x)
print(f"Prediction for x={test_x}: y={predicted_y}")
```
